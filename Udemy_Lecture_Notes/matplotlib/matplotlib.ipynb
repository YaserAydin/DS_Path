```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,9), dpi=100)

axes[0,0].plot(x,y)

axes[0,1].plot(x,y)

axes[1,0].plot(x,y)
axes[1,0].set_ylabel("Y Label 1,0")
axes[1,0].set_xlim(2,6)

axes[1,1].plot(a,b)
axes[1,1].set_title("Title 1,1")

fig.suptitle("Figure Level",fontsize=16);

plt.tight_layout() # grafik eksenlerinin üst üste binmesine engel olur
# fig.subplots_adjust(wspace=0.9) # grafik arası mesafeleri manuel olarak ayarlayabilirsiniz.Default değerde otomatik mesafe açar 

fig.savefig("new_subplots.png", bbox_inches= "tight")
#fig.set_figwidth(10)  alt figürleri genişletmeye yarar.
```

Legend

```python
x = np.linspace(0,11,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, label="X vs X")
ax.plot(x**2, label= "X vs X^2")
ax.legend(loc = (0.4,0.85));
```

Colors and Styles

```python
x = np.linspace(0,11,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label= "X & X", color= "#e6c017", lw= 4,ls= ":") 
# RGB HEX kod kullanılarak yapılabilir  
ax.plot(x, x+1, label="X & X+1", color = "darkblue", linewidth =1.9, linestyle= "-.") 
# possible linestyle options "--", "-", "-.", ":", "steps"
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x, x+1, label="X & X+1", color = "darkblue", linewidth =5, linestyle= "-.") 
lines[0].set_dashes([1,2,10,2])
# set_dashes özel çizgi şekli yapmanıızı sağlar. Örnektede 1 nokta dolu 2 nokta boş 10 nokta dolu 2 nokta boş anlamına geliyor
```
