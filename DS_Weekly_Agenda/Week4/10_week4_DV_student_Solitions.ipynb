```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

```python
sns.set_style(style= "darkgrid")
```

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

### Pandas plot method

```python
df.plot.bar();
```

## Matplotlib method

```python
index_axis = np.arange(len(index))
  
plt.bar(index_axis - 0.2, income, 0.4, label = 'income')
plt.bar(index_axis + 0.2, expense, 0.4, label = 'expense')
  
plt.xticks(index_axis, index)

plt.legend()
plt.show()
```

# 3. Stack the bars

```python
df.plot.bar(stacked = True); # stacked yığın haline getirir.
```

# 3. Rotate the labels and set figsize

```python
# 1.method
df.plot.bar(stacked= True, rot= 0, figsize=(9,6));
```

```python
# 2. method
fig, ax = plt.subplots(figsize = (10,5))
df.plot.bar(stacked = True, ax = ax)
plt.xticks(rotation = 360)
plt.show()
```

# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"
# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
df["labels"] = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
```

```python
b = df.plot.bar(figsize=(9,6), rot=0)
b.set_index("labels")
```

```python
df.set_index("labels", inplace = True)
```

```python
fig, ax = plt.subplots(figsize = (9,6))
df.plot.bar(ax = ax)
plt.xticks(rotation = 360)

for index, value in enumerate(income):
    plt.text(index, value, str(value), ha = "right")
for index, value in enumerate(expense):
    plt.text(index, value, str(value), ha = "left")

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (16,9))

sns.violinplot(data = tips, x = "size", y = "total_bill", ax = ax[0,0])
sns.boxplot(x = "size", y = "total_bill", data = tips, ax = ax[0,1])
tips.plot.scatter(x = "tip", y = "total_bill", c = "size", cmap = "viridis", ax = ax[1,0])
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes, ax = ax[1,1])
```
